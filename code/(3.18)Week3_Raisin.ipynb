```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

file_path ="/Users/yunseongmun/Documents/Ai_Lecture/data/Raisin_Dataset/Raisin_Dataset.xlsx"

df = pd.read_excel(file_path)
df
```

```python
# 'class' 열의 고유 값과 그 개수 확인
class_counts = df['Class'].value_counts()

# 결과 출력 (class가 골고루 분포돼있는게 좋은 데이터)
print(class_counts)

sns.countplot(data = df, x='Class')

plt.xlabel("Kecimen or Besni")
plt.ylabel("Count")
plt.title("Class")
```

```python
x=df.drop('Class',axis=1)
x
```

```python
y=df['Class']
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# Build a logistic regression classifier and predict

cli_lr = LogisticRegression(random_state=0)
cli_lr.fit(x_train, y_train)

pred_lr = cli_lr.predict(x_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```
